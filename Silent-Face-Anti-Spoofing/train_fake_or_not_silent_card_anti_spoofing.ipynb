```python
import os
import cv2

def crop_and_resize_images(input_dir, output_dir_org, output_dir_2_5x, output_dir_2x):
    # Create output directories if they don't exist
    os.makedirs(output_dir_2_5x, exist_ok=True)
    os.makedirs(output_dir_2x, exist_ok=True)
    os.makedirs(output_dir_org, exist_ok=True)

    # Iterate through each .txt file in the input directory
    for file_name in os.listdir(input_dir):
        if file_name.endswith('.txt'):
            # Get the base name without extension
            base_name = os.path.splitext(file_name)[0]
            txt_file_path = os.path.join(input_dir, file_name)
            jpg_file_path = os.path.join(input_dir, base_name + '.jpg')

            # Check if corresponding .jpg file exists
            if os.path.exists(jpg_file_path):
                # Read the image
                image = cv2.imread(jpg_file_path)

                # Read the coordinates from the .txt file
                with open(txt_file_path, 'r') as f:
                    x1, y1, w, h, _ = map(float, f.readline().split())

                # Calculate expanded dimensions
                def get_expanded_crop(image, x1, y1, w, h, expand_ratio):
                    center_x = x1 + w / 2
                    center_y = y1 + h / 2
                    new_w = w * expand_ratio
                    new_h = h * expand_ratio
                    new_x1 = int(center_x - new_w / 2)
                    new_y1 = int(center_y - new_h / 2)
                    new_x2 = int(center_x + new_w / 2)
                    new_y2 = int(center_y + new_h / 2)

                    # Ensure coordinates are within image boundaries
                    new_x1 = max(0, new_x1)
                    new_y1 = max(0, new_y1)
                    new_x2 = min(image.shape[1], new_x2)
                    new_y2 = min(image.shape[0], new_y2)

                    return image[new_y1:new_y2, new_x1:new_x2]

                # Crop and resize the image for original
                cropped_img = get_expanded_crop(image, x1, y1, w, h, 1)
                cv2.imwrite(os.path.join(output_dir_org, base_name + '.jpg'), cropped_img)

                # Crop and resize the image for 2.5x expansion
                cropped_img_2_5x = get_expanded_crop(image, x1, y1, w, h, 2.5)
                cv2.imwrite(os.path.join(output_dir_2_5x, base_name + '_2.5x.jpg'), cropped_img_2_5x)

                # Crop and resize the image for 2x expansion
                cropped_img_2x = get_expanded_crop(image, x1, y1, w, h, 2.0)
                cv2.imwrite(os.path.join(output_dir_2x, base_name + '_2x.jpg'), cropped_img_2x)

                print(f'Processed {base_name}.jpg')

# Directories
input_dir = '/mnt/ssd/ekyc_myanmar/project_fake_profile/fake_profile'
output_dir_org = '/mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing/datasets/rgb_image/1_80x80/1'
output_dir_2_5x = '/mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing/datasets/rgb_image/2.5_80x80/1'
output_dir_2x = '/mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing/datasets/rgb_image/2_80x80/1'
crop_and_resize_images(input_dir, output_dir_org, output_dir_2_5x, output_dir_2x)

input_dir = '/mnt/ssd/ekyc_myanmar/project_fake_profile/not_fake_profile'
output_dir_org = '/mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing/datasets/rgb_image/1_80x80/0'
output_dir_2_5x = '/mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing/datasets/rgb_image/2.5_80x80/0'
output_dir_2x = '/mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing/datasets/rgb_image/2_80x80/0'
crop_and_resize_images(input_dir, output_dir_org, output_dir_2_5x, output_dir_2x)

input_dir = '/mnt/ssd/ekyc_myanmar/project_fake_profile/silent_anti_tmp/0'
output_dir_org = '/mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing/datasets/rgb_image_test/1_80x80/0'
output_dir_2_5x = '/mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing/datasets/rgb_image_test/2.5_80x80/0'
output_dir_2x = '/mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing/datasets/rgb_image_test/2_80x80/0'
crop_and_resize_images(input_dir, output_dir_org, output_dir_2_5x, output_dir_2x)

input_dir = '/mnt/ssd/ekyc_myanmar/project_fake_profile/silent_anti_tmp/1'
output_dir_org = '/mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing/datasets/rgb_image_test/1_80x80/1'
output_dir_2_5x = '/mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing/datasets/rgb_image_test/2.5_80x80/1'
output_dir_2x = '/mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing/datasets/rgb_image_test/2_80x80/1'
crop_and_resize_images(input_dir, output_dir_org, output_dir_2_5x, output_dir_2x)
```

```python
%cd /mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing


import os
import cv2
import onnxruntime as ort
import numpy as np
from PIL import Image
import torchvision.transforms as vstrans
from src.data_io import transform as trans


class AntiSpoofPredict:
    def __init__(self, device_id):
        # ONNX Runtime automatically handles CPU and GPU devices, but we can set the preferred device
        self.device = 'cuda' if ort.get_device() == 'GPU' and torch.cuda.is_available() else 'cpu'
        self.session = ort.InferenceSession(model_path)

    def predict(self, img, model_path):
        # Preprocessing
        test_transform = vstrans.Compose([
            vstrans.ToPILImage(),
            vstrans.Resize((80, 80)),  # Resize to 80x80
            vstrans.ToTensor(),
        ])
        img = test_transform(img).unsqueeze(0).numpy()  # Convert to numpy array with batch dimension

        # Run inference
        input_name = self.session.get_inputs()[0].name
        result = self.session.run(None, {input_name: img})[0]  # Get the first output
        print(result)

        # Apply softmax to get probabilities
        probabilities = np.exp(result) / np.sum(np.exp(result), axis=1, keepdims=True)
        predicted_class = np.argmax(probabilities, axis=1)  # Get the index of the max probability

        return predicted_class, probabilities

    def predict_folder(self, img_folder, model_path):
        results = {}
        count_total = 0
        count_ones = 0  # Counter for the number of 1s in predictions
        for img_name in os.listdir(img_folder):
            if img_name.endswith(('.jpg', '.jpeg', '.png')):
                img_path = os.path.join(img_folder, img_name)
                img = cv2.imread(img_path)
                predicted_class, probabilities = self.predict(img, model_path)
                results[img_name] = predicted_class
                count_total += 1
                if predicted_class == 1:
                    count_ones += 1
                print(f"Processed {img_name}: {predicted_class} {probabilities}")

        print(f"Total number of 'FAKE's in predictions: {count_ones}/{count_total}")
        return results


if __name__ == "__main__":
    # Replace these with your actual image folder path and model path
    img_folder = '/mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing/datasets/rgb_image_test/1_80x80/1'
    model_path = '/mnt/ssd/ekyc_myanmar/project_fake_profile/Silent-Face-Anti-Spoofing/ProfileAntiSpoof__2024-08-27-16-45_1_80x80_MiniFASNetV2SE.onnx'

    # Initialize the AntiSpoofPredict class
    predictor = AntiSpoofPredict(device_id=0)

    # Run predictions on the folder of images
    predictions = predictor.predict_folder(img_folder, model_path)
```
